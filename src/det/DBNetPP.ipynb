# DBNet++ Inference

Run the following cell to execute the inference script on a single test image.
The script will display the result inline

Parameters:
- `--image_path`: Path to a single image
- `--min_area`: Filter out small noise (default 10)
- `--unclip_ratio`: Controls expansion of polygon (default 1.2)
- `--box_thresh`: Confidence threshold (default 0.6)
- `--visualize`: Show the full image with detections
- `--visualize_crops`: Show each cropped text region
- `--preprocess`: Run AI preprocessing (rembg + OpenCV)
- `--model_path`: Path to the model checkpoint
